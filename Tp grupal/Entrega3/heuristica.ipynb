```python
import csv
capacidad_x_centro_ruta = 'capacidades_reducido.csv'
distancia_vot_centro_ruta = 'constantes_reducido.csv'
CAP_MIN = 30
```

```python
capacidad_x_centro = {}
with open(capacidad_x_centro_ruta, "r") as f:
    capacidad_x_centro_csv = csv.reader(f)
    principio = True
    for id_centro, capacidad in capacidad_x_centro_csv:
        if principio:
            principio = False
            continue
        capacidad_x_centro[int(id_centro)] = int(capacidad)

capacidad_x_centro
```

```python
distancia_vot_centro = {}
with open(distancia_vot_centro_ruta, "r") as f:
    distancia_vot_centro_csv = csv.reader(f)
    principio = True
    for id_votante, id_centro, distancia in distancia_vot_centro_csv:
        if principio:
            principio = False
            continue
        distancia_vot_centro[(int(id_centro), int(id_votante))] = float(distancia)

distancia_vot_centro
```

```python
cant_votantes = int(len(distancia_vot_centro.keys())/len(capacidad_x_centro.keys()))

distancia_centros = {}
for centro in capacidad_x_centro:
    distancia_por_votante = {}
    for id_votante in range(1,cant_votantes):
        distancia_por_votante[id_votante] = distancia_vot_centro[(centro, id_votante)]
    distancia_centros[centro] = distancia_por_votante

distancia_centros
```

```python
mas_cercanos = {}
votante_asignado = set()
for centro in distancia_centros:
    result = []
    votantes_ordenados = sorted(distancia_centros[centro].items(), key=lambda kv: kv[1])
    i = 0
    while i < capacidad_x_centro[centro] and len(votante_asignado) != cant_votantes:
        votante = votantes_ordenados[i][0]
        i += 1
        if votante in votante_asignado:
            continue
        result.append(votante)
        votante_asignado.add(votante)
    if len(result) < CAP_MIN:
        for votante in result:
            votante_asignado.remove(votante)
    else:
        mas_cercanos[centro] = result

for centro in mas_cercanos:
    print(len(mas_cercanos[centro]))
```
